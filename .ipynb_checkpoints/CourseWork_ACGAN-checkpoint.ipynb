```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import PIL
import time
from IPython.display import display
```

```python
from tensorflow.keras.layers import *
from tensorflow.keras import Model
from tensorflow.keras.losses import *
from tensorflow.keras import activations
```

```python
class ACGAN(Model):
    def __init__(self):
        super(ACGAN,self).__init__()
        self.latent_dim = 100
        self.classes = 10
        self.input_size = (32,32,3)
        self.g_opt = tf.keras.optimizers.Adam(1e-4)
        self.d_opt = tf.keras.optimizers.Adam(1e-4)
        self.generator = self.build_generator_model()
        self.discriminator = self.build_discriminator_model()
    def build_generator_model(self):
        x0 = Input((self.latent_dim,))
        x1 = Input((self.classes,))
        x = concatenate([x0, x1])
        x = Dense(4*4*256, use_bias=False, input_shape=(self.latent_dim,))(x)
        x = BatchNormalization(name='G_BN_0')(x)
        x = LeakyReLU()(x)
        x = Reshape((4, 4, 256))(x)
        for num_filter in [128,64,32]:
            x = Conv2DTranspose(num_filter, (4, 4), strides=(2, 2), padding='same', use_bias=False)(x)
            x = BatchNormalization(name='G_BN_'+str(num_filter))(x)
            x = LeakyReLU()(x)
        x = Conv2D(3, (3, 3), strides=(1, 1), padding='same', activation='tanh')(x)
        return Model(inputs=[x0, x1], outputs=x)
    def build_discriminator_model(self):
        x_input = Input(self.input_size)
        x = Conv2D(64, (3, 3), strides=(2, 2), padding='same')(x_input)
        x = LeakyReLU()(x)
        for i in range(0,3):
            x = Conv2DTranspose(128, (3, 3), strides=(2, 2), padding='same', use_bias=False)(x)
            x = BatchNormalization(name='D_BN_'+str(i))(x)
            x = LeakyReLU()(x)
        x = Flatten()(x)
        x0 = Dense(1)(x)
        x1 = Dense(10)(x)
        return Model(inputs=x_input, outputs=[x0, x1])

    def g_loss(self,generated_output, labels):
        out_d, out_c = generated_output
        loss_d = tf.nn.sigmoid_cross_entropy_with_logits(tf.ones_like(out_d), out_d)
        loss_c = tf.nn.softmax_cross_entropy_with_logits(labels, out_c)
        return loss_d+loss_c
    def d_loss(self,real_output, generated_output, labels):
        real_out_d, real_out_c = real_output
        real_loss_d = tf.nn.sigmoid_cross_entropy_with_logits(labels=tf.ones_like(real_out_d), logits=real_out_d)
        real_loss_c = tf.nn.softmax_cross_entropy_with_logits(labels, real_out_c)
        real_loss = real_loss_d + real_loss_c

        generated_out_d, generated_out_c = generated_output
        generated_loss = tf.nn.sigmoid_cross_entropy_with_logits(labels=tf.zeros_like(generated_out_d), logits=generated_out_d)
        total_loss = real_loss + generated_loss

        return total_loss

    def train(self, images, labels, batch_size):
        noise = tf.random.normal([batch_size, self.latent_dim])
        with tf.GradientTape() as g_tape, tf.GradientTape() as d_tape:
            generated_images= self.generator([noise, labels], training=True)

            real_output = self.discriminator(images, training=True)
            generated_output = self.discriminator(generated_images, training=True)

            G_loss = self.g_loss(generated_output, labels)
            D_loss = self.d_loss(real_output, generated_output, labels)

        gradients_of_generator = g_tape.gradient(G_loss, self.generator.trainable_variables)
        gradients_of_discriminator = d_tape.gradient(D_loss, self.discriminator.trainable_variables)

        self.g_opt.apply_gradients(zip(gradients_of_generator, self.generator.trainable_variables))
        self.d_opt.apply_gradients(zip(gradients_of_discriminator, self.discriminator.trainable_variables))
```

```python
def img_to_float(img):
    return (np.float32(img)-127.5)/127.5
def img_to_uint8(img):
    return np.uint8(img*127.5+128).clip(0, 255)
```

```python
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.cifar10.load_data()
train_img_f32 = img_to_float(train_images)
BUFFER_SIZE = train_img_f32.shape[0]
BATCH_SIZE = 25
train_dataset_y = tf.data.Dataset.from_tensor_slices(train_labels[:, 0]).map(lambda y: tf.one_hot(y, 10))
train_dataset_x = tf.data.Dataset.from_tensor_slices(train_img_f32)
train_dataset = tf.data.Dataset.zip((train_dataset_x, train_dataset_y)).shuffle(BUFFER_SIZE).batch(BATCH_SIZE)
```

```python
acgan = ACGAN()
num_examples_to_generate = 20
latent_dim = 100

random_vector_for_generation = tf.random.normal([num_examples_to_generate,
                                                 latent_dim])
condition_vector_generation = tf.one_hot(list(range(10))+list(range(10)), 10)
```

```python
def start_epoch(beg,end):
    for epoch in range(beg,end):
        start_time = time.time()
        for images, labels in train_dataset:
            acgan.train(images, labels, BATCH_SIZE)
        fake = acgan.generator([random_vector_for_generation, condition_vector_generation], training=False)
        fake_concat = np.transpose(img_to_uint8(fake), [1,0,2,3]).reshape((32,-1,3))
        print("epoch {}, {} secs".format(epoch, time.time()-start_time))
        display(PIL.Image.fromarray(fake_concat))
```

```python
for epoch in range(15):
    start_time = time.time()
    for images, labels in train_dataset:
        acgan.train(images, labels, BATCH_SIZE)
    fake = acgan.generator([random_vector_for_generation, condition_vector_generation], training=False)
    fake_concat = np.transpose(img_to_uint8(fake), [1,0,2,3]).reshape((32,-1,3))
    print("epoch {}, {} secs".format(epoch, time.time()-start_time))
    display(PIL.Image.fromarray(fake_concat))
```

```python
start_epoch(16,45)
```
